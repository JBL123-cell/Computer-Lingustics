```python
import wikipediaapi as wikiapi
```

```python
wiki = wikiapi.Wikipedia('ru')
page = wiki.page('Галактика')
```

Корпус из текстовых документов был создан при помощи Wikipedia-API. Из страницы 'Галактика' в Википедии были изьяты ссылки на другие статьи по схожей тематике. Внизу есть перечень того, о чем в последующем были собраны документы. Это, к примеру, 'Звездный поток', 'Планета', 'Небо' и т.д

```python
links = page.links
```

```python
links
```

Всего в статье о Галактике были найдены 294 ссылок

```python
len(links)
```

Далее для каждой ссылки находилась статья из Википедии, описание к которой добавлялось в созданный список 'doc'. Статьи, в которых содержались года были пропущены, так как summary у них одинаковое и смысловой нагрузки не несет.

```python
doc = []
for title in sorted(links.keys()):
    if any(ch.isdigit() for ch in title):
        continue
    else:
        page_1 = wiki.page(title)
        if page_1.exists():
            doc.append(page_1.summary)
```

Посмотрим суммарное количество слов в списке

```python
a = 0
for st in doc:
    a += int(len(st.split()))
print(a)
```

Узнаем среднее количество слов на документ (общее количество слов делим на количество документов)

```python
35431/292
```

Вычислим медиану количества слов в документах

```python
from statistics import median
med = []
for st in doc:
    med.append(int(len(st.split())))
median(med)
```

Посчитаем количество документов

```python
b = 0
for dc in doc:
    b += 1
print(b)
```

Далее преобразуме список doc в словарь data, который записываем  в файл json

```python
import json
data = {}
for i in range(len(doc)):
    data[i] = {'text': doc[i]}
```

```python
with open('fl.json', 'w', encoding='utf-8') as file:
    json.dump(data, file, indent=4, ensure_ascii=False)
```
