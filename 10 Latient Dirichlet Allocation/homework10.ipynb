```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
import pymorphy2
import pandas as pd
import re
analyzer = pymorphy2.MorphAnalyzer()
```

## Предобработка текста

```python
new_corpus = pd.read_csv('texts.csv', encoding='utf-8')
```

```python
new_corpus.head()
```

```python
stopwords_list = stopwords.words("russian") + ["это", "который", "наш", "мочь", "год", 
                                            "такой", "знать", "мы", "свой", "один", "другой", "хотеть",
                                            "человек", "всё", "весь", "очень", "думать", "нужно",
                                            "большой", "время", "использовать", "говорить", "сказать",
                                            "иметь", "сделать", "первый", "каждый", "день", "её", "ваш",
                                            "стать", "больший", "ваше", "день", "самый", "понять",
                                            "просто", "ещё", "проблема", "также", "например", "июнь", "январь",
                                              "ферваль", "март", "апрель", "май", "июль", "август", "сентябрь", "октябрь",
                                              "ноябрь", "декабрь"]
```

```python
corpus = new_corpus['text'].to_list()
```

```python
def _change(list_of_texts):
    for i in range(len(list_of_texts)):
        list_of_texts[i] = re.sub(r"[^а-яА-Я]",' ', list_of_texts[i])
        list_of_texts[i] = re.sub(r"\xa0",' ', list_of_texts[i])
        tokens = word_tokenize(list_of_texts[i])    
        tokens = [token.lower() for token in tokens if token not in stopwords_list]
        list_of_texts[i] = [word for word in tokens if word not in stopwords_list]
    return list_of_texts

def _speech(list_of_texts):
    all_lemma = []
    for i in range(len(list_of_texts)):
        one_lemma = []
        for word in list_of_texts[i]:
            if analyzer.parse(word)[0].tag.POS not in ('NPRO', 'PREP', 'CONJ', 'PRCL', 'INTJ'): 
                if analyzer.parse(word)[0].normal_form not in stopwords_list:
                    one_lemma.append(analyzer.parse(word)[0].normal_form)
        all_lemma.append(one_lemma)
    return all_lemma
```

```python
second_corpus_1 = _change(corpus)
second_corpus = _speech(corpus)
```

```python
second_corpus[0][:10]
```

## LDA

```python
import gensim.corpora as corpora
import pyLDAvis.gensim_models
import pyLDAvis
import pickle
```

Перейдем к созданию корпуса и построению модели

```python
id2word = corpora.Dictionary(corpus)
corpus_r = [id2word.doc2bow(cor) for cor in second_corpus]
corpus_r[0][:10]
```

Построим модель для пяти тем, так как начиная с 6-ой темы круги на графике начинают пересекаться

```python
from gensim.models import LdaMulticore
lda_model = LdaMulticore(corpus=corpus_r, id2word=id2word, num_topics=4)
```

```python
lda_model.print_topics()
```
